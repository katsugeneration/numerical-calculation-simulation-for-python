```python
%matplotlib inline
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
G = 9.80665
h = 0.01
```

```python
def f(param, t):
    return [param[1], -G]
```

```python
param = [100, 0]
t = np.arange(0, 4.53, h)
values = odeint(f, param, t)
plt.plot(t, values[:, 0])
plt.show()
```
