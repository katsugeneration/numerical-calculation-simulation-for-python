```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
G = 9.80665
```

```python
h = 0.01
```

```python
def freefall(t, v, x):
    tlist = [t]
    xlist = [x]

    while x >= 0:
        v -= G*h
        x += v*h
        t += h
        tlist.append(t)
        xlist.append(x)

    plt.plot(tlist, xlist)
    plt.show()
```

```python
t = 0
v = 0.0
x = 100.0

freefall(t, v, x)
```

```python
t = 0
v = 100.0
x = 0.0

freefall(t, v, x)
```

```python
def retrofire(t, v, x0, a, tf):
    x = x0
    tlist = [t]
    xlist = [x]

    while x >= 0 and x <= x0:
        v -= (G - (a if t >= tf else 0)) * h
        x += v*h
        t += h
        tlist.append(t)
        xlist.append(x)

    plt.plot(tlist, xlist)
    plt.show()
```

```python
t = 0
v = 0.0
x = 100.0

retrofire(t, v, x, 1.5 * G, 2.62)
```

```python
t = 0
v = 0.0
x = 100.0

retrofire(t, v, x, 1.5 * G, 1.0)
```
