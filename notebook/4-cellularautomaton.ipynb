```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def make_rule(R, no):
    rule_string = format(no, f'0{R}b')
    return [int(r) for r in reversed(rule_string)]
```

```python
def iteration(rule, cells):
    next_cells = np.zeros(len(cells), dtype=int)
    for i in range(1, len(cells) - 1):
        next_cells[i] = rule[cells[i-1] * 4 + cells[i] * 2 + cells[i+1]]
    return next_cells
```

```python
R = 8
no = 18
MAXIT = 256
N = 256
rule = make_rule(R, no)
print('rule:', rule)
cells = np.zeros(N+2, dtype=int)
cells[50] = 1
output = []

for _ in range(MAXIT):
    output.append(cells)
    cells = iteration(rule, cells)

plt.imshow(output)
plt.show()
```
