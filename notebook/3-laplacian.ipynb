```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
def iteration(u, N, M):
    u_next = np.zeros((N, M))
    for i in range(1, N - 1):
        for j in range(1, M - 1):
            u_next[i][j] = (u[i-1][j] + u[i+1][j] + u[i][j-1] + u[i][j+1]) / 4
    u[1:N-1, 1:M-1] = u_next[1:N-1, 1:M-1]
```

```python
LIMIT = 1000
N = 101
M = 101

u = np.zeros((N, M))
for i in range(M):
    u[0][i] = np.sin(2 * np.pi * i / (M - 1))

for _ in range(LIMIT):
    iteration(u, N, M)
```

```python
X = np.arange(0, N)
Y = np.arange(0, M)
X, Y = np.meshgrid(X, Y)
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, u, cmap=cm.coolwarm)
plt.show()
```

```python
LIMIT = 1000
N = 101
M = 101

u = np.zeros((N, M))
for i in range(M):
    u[0][i] = np.sin(2 * np.pi * i / (M - 1))
    u[N-1][i] = np.cos(2 * np.pi * i / (M - 1))

for _ in range(LIMIT):
    iteration(u, N, M)

X = np.arange(0, N)
Y = np.arange(0, M)
X, Y = np.meshgrid(X, Y)
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, u, cmap=cm.coolwarm)
plt.show()
```

```python
def iteration_specific(u, N, M):
    u_next = np.zeros((N, M))
    for i in range(int(N / 4), int((N - 1) * 3 / 4)):
        for j in range(1, int((M - 1) / 4)):
            u_next[i][j] = (u[i-1][j] + u[i+1][j] + u[i][j-1] + u[i][j+1]) / 4
    for i in range(1, N - 1):
        for j in range(int((M - 1) / 4), int((M - 1) * 3 / 4)):
            u_next[i][j] = (u[i-1][j] + u[i+1][j] + u[i][j-1] + u[i][j+1]) / 4
    for i in range(int(N / 4), int((N - 1) * 3 / 4)):
        for j in range(int((M - 1) * 3 / 4), M - 1):
            u_next[i][j] = (u[i-1][j] + u[i+1][j] + u[i][j-1] + u[i][j+1]) / 4
    u[1:N-1, 1:M-1] = u_next[1:N-1, 1:M-1]
```

```python
LIMIT = 1000
N = 101
M = 101

u = np.zeros((N, M))
for i in range(M):
    u[0][i] = np.sin(2 * np.pi * i / (M - 1))

for _ in range(LIMIT):
    iteration_specific(u, N, M)

X = np.arange(0, N)
Y = np.arange(0, M)
X, Y = np.meshgrid(X, Y)
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, u, cmap=cm.coolwarm)
plt.show()
```
